```python
from pyspark.sql import functions as F
import json
```

```python
url = "jdbc:postgresql://yourpgdbserver.postgres.database.azure.com:5432/adventureworks"
properties = {
    "user": "admins-aaaorg",
    "password": password
}
```

```python
def convert_array_to_string(array):    
    quoted_array = ["'{}'".format(element) for element in array]
    result = ",".join(quoted_array)
    return result
```

```python

filterParams = convert_array_to_string(schemastoload)
print(filterParams)
```

```python
query = "(SELECT table_schema, table_name FROM information_schema.tables WHERE table_schema  in (" + filterParams + ") and table_type = 'BASE TABLE' ORDER BY table_schema ASC, table_name ASC) AS t"
```

```python
# Execute the query and load the result into a DataFrame
df = spark.read.jdbc(url=url, table=query, properties=properties)
```

```python
# Convert the 'tables' column to a JSON string
#json_df = grouped_df.withColumn("tables", F.to_json("tables"))
# Collect the result into a list of Row objects
#result = json_df.collect()
dictionary=[]
for row in df.collect():
    dictionary.append(row.asDict())
print(dictionary)
with open('/lakehouse/default/Files/PgDictionary/PgDictString.txt', 'w') as f:
    json.dump(dictionary, f)
```
