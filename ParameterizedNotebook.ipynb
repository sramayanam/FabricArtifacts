```python
tablename = "document"
schemaname = "production"
password = ""
```

```python
sourcetablename = schemaname + "." + tablename
```

```python
df = spark.read \
    .format("jdbc") \
    .option("url", "jdbc:postgresql://yourpgdbserver.postgres.database.azure.com:5432/adventureworks") \
    .option("dbtable", sourcetablename  ) \
    .option("user", "admins-aaaorg") \
    .option("password", password) \
    .load()
```

```python
location = "Tables/"+ schemaname + "_" + tablename
```

```python
df.write.mode("overwrite").format("delta").save(location)
```
